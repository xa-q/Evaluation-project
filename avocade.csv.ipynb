```python
x=df1.drop('AveragePrice',axis=1)
y=df1['AveragePrice']
```

```python
import statsmodels.api as sm
from scipy import stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(d):
    vif=pd.DataFrame()
    vif['Variable']=d.columns
    vif['vif factor']=[variance_inflation_factor(d.values,i) for i in range(d.shape[1])]
    return (vif)
```

```python
x=x.drop('Total Bags', axis=1)
```

```python
x.head()
```

```python
calc_vif(x)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
le=LinearRegression()
msem=0
r2m=0
rsb=0
for i in range(5000):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=3,random_state=i)
    le.fit(x_train,y_train)
    pred=le.predict(x_test)
    score=le.score(x_train,y_train)
    score2=le.score(x_test,y_test)
    r2=r2_score(y_test,pred)
    mse=mean_squared_error(y_test,pred)
    if r2>r2m:
        r2m=r2
        rsb=i
        msem=mse
print('Max r2 score:',r2m)
print('mse:',msem)
print('random state:',rsb)
print('Train & Test scores:',score,score2)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=3,random_state=3750)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
models=[LinearRegression(),SVR(),RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor()]
for m in models:
    m.fit(x_train,y_train)
    predm=m.predict(x_test)
    print(f'Metrics of {m}:')
    print('Score:',m.score(x_train,y_train))
    
    print('mse:',mean_squared_error(y_test,predm))
    print('r2 score:',r2_score(y_test,predm))
    print('\n')
```

```python
from sklearn.model_selection import cross_val_score
for m in models:
    cvscore=cross_val_score(m,x,y,cv=5)
    print(f'cvscore of {m}:')
    print('Mean cv score:',cvscore.mean())
    print('Std:',cvscore.std())
    print('\n')
```

```python
gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
score=gbr.score(x_train,y_train)
score2=gbr.score(x_test,y_test)
pred=gbr.predict(x_test)
r2=r2_score(y_test,pred)
mse=mean_squared_error(y_test,pred)
print('percent of training data :',score)
print('percent of test data :',score2)
print('mean squared error of model:',mse)
print('r2 score:',r2)
```

```python
xa=df1.drop('region',axis=1)
yb=df1['region']
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(d):
    vif=pd.DataFrame()
    vif['Variable']=d.columns
    vif['vif factor']=[variance_inflation_factor(d.values,i) for i in range(d.shape[1])]
    return (vif)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xa_scaled=scaler.fit_transform(xa)
xa_scaled
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
```

```python
dtc=DecisionTreeClassifier()
macc_score=0
rs=0
for i in range(100):
    x_train,x_test,y_train,y_test=train_test_split(xa,yb,test_size=0.25,random_state=i)
    dtc.fit(x_train,y_train)
    predr=dtc.predict(x_test)
    acc=accuracy_score(y_test,predr)
    if acc>macc_score:
        macc_score=acc
        rs=i
print('Max acuuracy  score:',macc_score)
print('Best random state is :', rs)
```

```python
x_train,x_test,y_train,y_test=train_test_split(xa,yb,test_size=0.25,random_state=93)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
rf=RandomForestClassifier(n_estimators=10)
rf.fit(x_train,y_train)
predr=rf.predict(x_test)
print('Metrics of Random Forest:')
print('Score:',accuracy_score(y_test,predr))
print('Confusion Matrix:',confusion_matrix(y_test,predr))
print('Classification Report:',classification_report(y_test,predr))
```

```python
bc=AdaBoostClassifier(n_estimators=100)
bc.fit(x_train,y_train)
predr=bc.predict(x_test)
print('Metrics of AdaBoostClassifier:')
print('Score:',accuracy_score(y_test,predr))
print('Confusion Matrix:',confusion_matrix(y_test,predr))
print('Classification Report:',classification_report(y_test,predr))
```

```python
gc=GradientBoostingClassifier()
gc.fit(x_train,y_train)
predr=gc.predict(x_test)
print('Metrics of GradientBoostingClassifier:')
print('Score:',accuracy_score(y_test,predr))
print('Confusion Matrix:',confusion_matrix(y_test,predr))
print('Classification Report:',classification_report(y_test,predr))
```
